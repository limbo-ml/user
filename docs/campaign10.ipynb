**Description**

Each image contains three groups of randomly-chosen distractor objects, with members of each group placed at random, but without overlap.

Note that, due to differences in size among distractors (ranging from a paint can to a massive oil tank) and random camera placement, only one or two groups may be visible in the actual image.

In previous campaigns, the barrels in a palette-load of six barrels banded together were labeled *barrel/55G/A* through *barrel/55G/F*.  Starting with this campaign, they are all simply labelled *barrel/55G*.
