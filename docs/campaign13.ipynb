**Description**

Each image contains individual type 48 containers surrounded by randomly-chosen distractor objects, viewed from above (it's the "Where's Waldo?" of uranium hexafluoride).  This data should be extremely challenging for models.

Note that, due to differences in size among distractors (ranging from a paint can to a massive oil tank) and random camera placement, the type 48 containers may not always be visible in the actual image.
